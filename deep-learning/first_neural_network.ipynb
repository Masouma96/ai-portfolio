```python
%pip install tensorflow
import tensorflow as tf

# Data
x = tf.constant([0.03, 0.19, 0.34, 0.46, 0.78, 0.81, 1.08, 1.18, 1.39, 1.60, 1.65, 1.90])
y = tf.constant([0.67, 0.85, 1.05, 1.0, 1.40, 1.5, 1.3, 1.54, 1.55, 1.68, 1.73, 1.6 ])

# Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,), activation="linear")
])

# Compile
model.compile(optimizer=tf.keras.optimizers.SGD(), loss=tf.keras.losses.mse)

# Train
model.fit(x, y, epochs=1000)
```

```python
%pip install matplotlib

import matplotlib.pyplot as plt
import numpy as np

x_values = np.linspace(0, 2, 100)
y_values = model.predict(x_values)

plt.scatter(x, y, label="Data Points")
plt.plot(x_values, y_values, color="red", label="Fitted Line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression with TensorFlow")
plt.legend()
plt.show()



```
